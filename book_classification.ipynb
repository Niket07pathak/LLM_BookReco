```python
category_mapping = {'Fiction' : "Fiction",
 'Juvenile Fiction': "Fiction",
 'Biography & Autobiography': "Nonfiction",
 'History': "Nonfiction",
 'Literary Criticism': "Nonfiction",
 'Philosophy': "Nonfiction",
 'Religion': "Nonfiction",
 'Comics & Graphic Novels': "Fiction",
 'Drama': "Fiction",
 'Juvenile Nonfiction': "Nonfiction",
 'Science': "Nonfiction",
 'Poetry': "Fiction"}

books["simple_categories"] = books["categories"].map(category_mapping)
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
candidate_labels = ["Fiction", "Nonfiction"]
```

```python
sequence = books.loc[books["simple_categories"] == "Fiction", "description"].reset_index(drop=True)[0]
```

```python
classifier(sequence, candidate_labels=candidate_labels)
```

```python
import numpy as np

max_index = np.argmax(classifier(sequence, candidate_labels)["scores"])
max_label = classifier(sequence, candidate_labels)["labels"][max_index]
max_label
```

```python
def classify_book(text,categories):
    max_index = np.argmax(classifier(text, categories)["scores"])
    max_label = classifier(text, categories)["labels"][max_index]
    return max_label
```

```python
isbns = []
pred_categories = []

missing = books.loc[books["simple_categories"].isna(), ["isbn13", "description"]].reset_index(drop=True)
```

```python
missing.shape
```

```python
for i in tqdm(range(missing.shape[0])):
    sequence = missing["description"][i]
    pred_categories+= [classify_book(sequence, candidate_labels)]
    isbns+= [missing["isbn13"][i]]
```

```python
missing_df = pd.DataFrame({"isbn13": isbns, "predicted_category": pred_categories})
missing_df.head()
```

```python
books = pd.merge(books, missing_df, on="isbn13", how="left")
books["simple_categories"] = books["simple_categories"].fillna(books["predicted_category"])
books = books.drop(columns=["predicted_category"])
```
